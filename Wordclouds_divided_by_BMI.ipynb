```python
import pandas as pd
diet = pd.read_csv('diet_above_avgBMI.csv')
diet.head()
```

```python
lst1=list(set(diet['diet_current']))
lst2=list(set(diet['eating_changes']))
lst=lst1 + lst2
lst_cleaned=[]
for l in lst:
    l=str(l).split()
    lst_cleaned.append(l)
lst_cleaned
    
```

```python
globalwords=[]
for l in lst_cleaned:
    for w in l:     
        w=w.strip(',').strip('.')
        if len(w) != 0 and w!= 'nan': globalwords.append(w)
len(globalwords)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
 
stop_words = set(stopwords.words('english'))
stop_words.update('.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}') 
filtered_words = [[w] for w in globalwords if w not in stop_words] 

len(filtered_words)
```

```python
filtered_words
```

```python
import csv
with open("above_words4cloud.csv",'wb') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(filtered_words)
```
